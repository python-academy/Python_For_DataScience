```python
hawks.shape
```

Some data science models require features to be formatted as an array.

```python
np.ravel(hawks['Wing'])
```

The "transpose" of an array switches the dimensions.

```python
np.transpose(hawks['Wing'])
```

It may also be useful to sort values of a feature from smallest to largest, or in alphabetical order.

```python
np.sort(hawks['Wing'])
```

```python
hawks[hawks['Wing'] <= 150]
```

```python
hawks[hawks['Wing'] == 100]
```

Mathematical operations from NumPy are useful for calulating new features in a dataset, or summary statistics.

```python
np.mean(hawks['Wing'])
```

```python
np.median(hawks['Wing'])
```

```python
hawks['Ratio'] = hawks['Wing'] / hawks['Weight']
np.mean(hawks['Ratio'])
```

## Try 2.8.3: matplotlib functions for data science.

```python
hawks
```

Plot all species using default options in plt.scatter().

```python
x = hawks['Wing']
y = hawks['Weight']

plt.scatter(x, y)
plt.xlabel('Wing', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.title('Wing vs. weight for all hawks', fontsize=16)
```

Group by species, then plot using custom options.

```python
CoopersHawk = hawks[hawks['Species'] == 'CH']
RedTailed = hawks[hawks['Species'] == 'RT']
SharpShinned = hawks[hawks['Species'] == 'SS']

x1 = CoopersHawk['Wing']
y1 = CoopersHawk['Weight']

x2 = RedTailed['Wing']
y2 = RedTailed['Weight']

x3 = SharpShinned['Wing']
y3 = SharpShinned['Weight']

plt.scatter(x1, y1, label="Cooper's", edgecolors='white', color='orange', marker='o')
plt.scatter(x2, y2, label='Red tailed', edgecolors='white', color='green', marker='v')
plt.scatter(x3, y3, label='Sharp-shinned', edgecolors='white', color='blue', marker='p')
plt.legend(fontsize=12)
plt.xlabel('Wing', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.title('Wing vs. weight by species', fontsize=16)
```
